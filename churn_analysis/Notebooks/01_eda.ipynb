```python
df.drop('customerID',axis=1, inplace=True)
```

```python
df ["TotalCharges"]= pd.to_numeric(df["TotalCharges"], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df[df['TotalCharges'].isnull()]
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df.isnull().sum()
```

```python
df[df['TotalCharges'].isnull()]
```

```python
df['SeniorCitizen']= df['SeniorCitizen'].astype('category')
```

```python
df['tenure_group']= pd.cut(df['tenure'], bins=[0,12,24,48,72], labels= ['0-1y', '1-2y', '2-4y', '4-6y'])
```

```python
df['high_value']=(df['MonthlyCharges']>df['MonthlyCharges'].median().astype(int))
```

```python
services=[ 'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies']
df['num_services']=df[services].apply(lambda x: (x == 'Yes').sum(), axis=1)
```

```python
df['avg_monthly_spend']= df['TotalCharges']/ (df['tenure'] + 1)
```

```python
## Churn distribution
sns.countplot(x= 'Churn', data= df)
plt.title('Churn Distribution')
plt.show()
```

```python
#Contract vs churn
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn by Contract Type')
plt.xticks(rotation=45)
plt.show()
```

```python
#Tenure group vs churn
sns.countplot(x='tenure_group', hue='Churn', data=df)
plt.title('Churn by Tenure group')
plt.show()
```

```python
#Monthly charges vs churn
sns.boxplot(x='Churn', y="MonthlyCharges", data=df)
plt.title('Monthly Charges vs Churn')
plt.show()
```

```python
#number of services vs churn
sns.boxplot(x='Churn', y='num_services', data=df)
plt.title('Number of services vs Churn')
plt.show()
```

```python
sns.boxplot(x='Churn', y='avg_monthly_spend', data=df)
plt.title('Average Monthly Spend vs Churn')
plt.show()
```
